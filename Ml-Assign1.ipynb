```python
from rdkit import Chem
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("qm9.csv",sep=",")
filtered_df = df[~df["smiles"].str.contains("S"and"Cl")]
#print(filtered_df)
A=filtered_df["u298_atom"]
U=np.array(A)*(1./627.5)
B=filtered_df["h298_atom"]
H=np.array(B)*(1./627.5)
C=filtered_df["g298_atom"]
G=np.array(C)*(1./627.5)
Na=np.zeros(133885)
str=np.array(filtered_df.smiles)
#count no. of atoms
for i in range(len(str)):
    mol = Chem.MolFromSmiles(str[i])
    mol=Chem.AddHs(mol)
    Na[i]=mol.GetNumAtoms()

N=np.array(Na)
plt1=sns.scatterplot(N,U)
plt1.set(xlabel ="Number of atoms", ylabel = "Atomization Energy U (hartees)", title ='U (hartrees) vs N')
plt1.xaxis.set_major_locator(MultipleLocator(3))
plt1.xaxis.set_minor_locator(AutoMinorLocator(1))
plt.show()
plt2=sns.histplot(U)
plt2.set(xlabel = "Atomization Energy U (hartees)")
plt.show()
plt3=sns.histplot(H)
plt3.set(xlabel = "Atomization Energy H (hartees)")
plt.show()
plt4=sns.histplot(G)
plt4.set(xlabel = "Atomization Energy G (hartees)")
plt.show()
```
