```python
from rdkit import Chem
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("qm9.csv",sep=",")
filtered_df = df[~df["smiles"].str.contains("S"and"Cl")]
#print(filtered_df)
A=df["u298_atom"]
U=np.array(A)*(1./627.5)
B=df["h298_atom"]
H=np.array(B)*(1./627.5)
C=df["g298_atom"]
G=np.array(C)*(1./627.5)
Na=np.zeros(133885)
str=np.array(df.smiles)
#count no. of atoms
for i in range(len(str)):
    mol = Chem.MolFromSmiles(str[i])
    mol=Chem.AddHs(mol)
    Na[i]=mol.GetNumAtoms()
N=np.array(Na)
plt1=sns.scatterplot(N,U)
plt1.set(xlabel ="Number of atoms", ylabel = "Atomization Energy U (hartees)", title ='U (hartrees) vs N')
plt.show()
plt2=sns.histplot(U)
plt2.set(xlabel = "Atomization Energy U (hartees)")
plt.show()
plt3=sns.histplot(H)
plt3.set(xlabel = "Atomization Energy H (hartees)")
plt.show()
plt4=sns.histplot(G)
plt4.set(xlabel = "Atomization Energy G (hartees)")
plt.show()

```
